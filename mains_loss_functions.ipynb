```python
import numpy as np
import matplotlib.pyplot as plt
```

Our loss function is a simple quadratic expression. The goal is to find the value of x that minimises this loss function.

$$
L(x) = (x-3)^2
$$

The derivative (gradient) of this loss function is.

$$
\frac{dL}{dx} = 2(x-3)
$$

```python
# Loss function we want to minimise with an x value
def loss(x):
    return x**4 - 6*x**2 + 4*x + 12

# Derivative / gradient
def gradient(x):
    return 4*x**3 - 12*x + 4

# Hessian
def hessian(x):
    return 12*x**2 - 12

# Initialize parameters
x = 0.5 # starting point
learning_rate = 0.03 # learning rate
num_iterations = 20 # number of iterations

# Lists to record the optimisation process
x_history = [x]
loss_history = [loss(x)]

# Gradient Descent Loop
for i in range(num_iterations):

    # Calculate gradient and Hessian
    grad = gradient(x)
    hess = hessian(x)

    # Scale the update of the gradient by the learning rate
    # *x = x - learning_rate * grad 
    # Scale update using the Hessian
    x = x - grad / hess
    # Store new x and loss values
    x_history.append(x)
    loss_history.append(loss(x))
```

```python
# Create a range of x values to plot the loss function.
x_range = np.linspace(-2.5, 2.5, 400)
y_range = loss(x_range)

# Plot the loss function.
plt.figure(figsize=(10, 6))
plt.plot(x_range, y_range, label="Loss Function: $(x-3)^2$", color="blue")

# Plot the iterative points on the loss function.
plt.scatter(x_history, loss(np.array(x_history)), color="red", zorder=5, label="Iterations")
plt.plot(x_history, loss(np.array(x_history)), color="red", linestyle="--")

# Annotate the points with iteration numbers.
for i, (xi, yi) in enumerate(zip(x_history, loss(np.array(x_history)))):
    plt.annotate(f"{i}", (xi, yi), textcoords="offset points", xytext=(0,5), ha='center')

plt.xlabel("x")
plt.ylabel("Loss")
plt.title("Gradient Descent on $(x-3)^2$: Iterative Updates")
plt.legend()
plt.show()

# Plot the convergence of the loss.
plt.plot(loss_history, marker='o')
plt.xlabel("Iteration")
plt.ylabel("Loss")
plt.title("Loss Convergence")
plt.show()
```
